```python
import torch
from d2l import torch as d2l
```

```python
img = d2l.plt.imread('../img/catdog.jpg')
h, w = img.shape[:2]
h, w
```

```python
def display_anchors(fmap_w, fmap_h, s):
    d2l.set_figsize(figsize=(6, 6)) 
    # 前两个维度上的值不影响输出 

    # me：fmap并不是由catdog.jpg经过卷积得到的，而是假设已经经过卷积计算后得到的，后续映射回catdog.jpg
    fmap    = torch.zeros((1, 7, fmap_h, fmap_w)) 
    # me: anchors中元素取值在0.0到1.0之间，表示锚框在图片中的相对位置
    anchors = d2l.multibox_prior(fmap, sizes=s, ratios=[1, 0.5, 2.0]) 
    print(anchors)
    bbox_scale = torch.tensor((w, h, w, h)) 
    d2l.show_bboxes(d2l.plt.imshow(img).axes, anchors[0] * bbox_scale)
```

```python
display_anchors(fmap_w=1, fmap_h=1, s=[0.9])
```
